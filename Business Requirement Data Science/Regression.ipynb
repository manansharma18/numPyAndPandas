## Conditions for Linear Regression :
1. Predict relation between target variable and independent linear variable
2. Data is free of missing value
3. There is a linear relationship between predictors and predictant
4. All predictors are independent of each other
5. Residuals or prediction errors are normally distributed. 


```python
import numpy as np
import pandas as pd
import sklearn 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from pylab import rcParams
from sklearn.preprocessing import scale 
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 10,8
```

```python
rooms = 2 * np.random.rand(100,1) + 3
rooms[1:10]
```

```python
price = 265 + 6*np.random.rand(100,1) + abs(np.random.randn(100,1))
price[0:5]
```

```python
linReg = LinearRegression()
roomPriceFit = linReg.fit(rooms,price)
print(roomPriceFit.coef_)
print(roomPriceFit.intercept_)
```

## Not checking for missing value as we have created the data. 

```python
print(roomPriceFit.predict(np.array([10,12]).reshape(-1,1)))
```

```python
plt.plot(rooms,price,'r^')
plt.show()
```

```python
print(linReg.score(rooms,price))
```

A negative 2 is only possible with linear regression when either the intercept or the slope are constrained so that the "best-fit" line (given the constraint) fits worse than a horizontal line.
